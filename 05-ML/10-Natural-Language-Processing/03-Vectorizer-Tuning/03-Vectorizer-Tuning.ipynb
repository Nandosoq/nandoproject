# Vectorizer Tuning

```python
import pandas as pd

data = pd.read_csv("reviews.csv")

data.head()
```

The dataset is made up of positive and negative movie reviews.

## Preprocessing

üëá Remove punctuation and lower case the text.

```python
import string 

def rem_punct(text):
    for punctuation in string.punctuation:
        text = text.replace(punctuation, '')
    return text

def low_case(text):
    text = text.lower() 
    return text
```

```python
data['clean_reviews'] = data['reviews'].apply(rem_punct)
data['clean_reviews'] = data['clean_reviews'].apply(low_case)
```

## Tuning

üëá Tune a vectorizer of your choice (or try both!) and a MultinomialNB model simultaneously.

```python
y = data['target']
```

```python
# Create Pipeline

from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer

from sklearn.naive_bayes import MultinomialNB

pipeline = Pipeline([
    ('tfidf', TfidfVectorizer()),
    ('nb', MultinomialNB()),
])
```

```python
# Set parameters to search (model and vectorizer)

# Set parameters to search
parameters = {
    'tfidf__ngram_range': ((1,1), (2,2)),
    'nb__alpha': (0.1, 0.2, 0.3),}
```

```python
# Perform grid search on pipeline
from sklearn.model_selection import GridSearchCV

grid_search = GridSearchCV(pipeline, parameters, n_jobs=-1, 
                           verbose=1, scoring = "accuracy", 
                           refit=True, cv=5)

grid_search.fit(data['clean_reviews'] ,y)

grid_search.best_params_
```
