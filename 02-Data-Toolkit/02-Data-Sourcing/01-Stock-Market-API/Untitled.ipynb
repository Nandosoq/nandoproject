```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib
```

```python
import requests

url = "http://iex.lewagon.com/stable/stock/aapl/chart/3m"
api_data = requests.get(url).json()
```

```python
apple_df  = pd.DataFrame(api_data)
```

Convert the date column to a datetime object

Set the date column as the index

```python
apple_df['date'] = pd.to_datetime(apple_df['date'], format='%Y-%m-%d')
apple_df.set_index('date',inplace=True)
```

First letâ€™s plot with only the values in the close column

Now we can make a plot with the values in the open, close, high, and low columns

```python
apple_df['close'].plot()
apple_df[['open','close','high','low']].plot(figsize=(12,4))
```
